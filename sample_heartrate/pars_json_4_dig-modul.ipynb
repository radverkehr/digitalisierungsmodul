## example for analysis
* merging heartrate and geolocation for viz

```python
recordedRoute=data['exercises'][0]['samples']['recordedRoute']
recordedRoute_df=pd.DataFrame.from_dict(recordedRoute)

recordedRoute_df.head()
```

```python
#prepare merge; cut miliseconds to match timestamp
recordedRoute_df.dateTime= recordedRoute_df.dateTime.str[:-1]
heartRate_df.dateTime= heartRate_df.dateTime.str[:-1]
```

```python
merge_df=recordedRoute_df.merge(heartRate_df)
merge_df.head()
```

```python
import geopandas as gpd

# create geodataframe
merge_gdf = gpd.GeoDataFrame(merge_df, geometry=gpd.points_from_xy(merge_df['longitude'], merge_df['latitude']), crs="4326")
```

```python
merge_gdf
```

```python
merge_gdf.explore("value", cmap="Blues", tiles="CartoDB positron")
```
